# Age Replacing
1. Mean
2. Median
3. Mode
4. User-defined Function

```python
sns.boxplot(x='Pclass',y='Age',data=titanic)
```

```python
def agge(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 30
    elif Pclass==3:
      return 28
  else:
    return Age
```

```python
titanic["Age"]=titanic[['Age','Pclass']].apply(agge,axis=1)
```
